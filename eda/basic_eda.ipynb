Las columnas Description (object) y CustomerID (float64) poseen nulos

```python
# Change type object or str to string to avoid future warnings
df["InvoiceNo"] = df["InvoiceNo"].astype("string")
df["StockCode"] = df["StockCode"].astype("string")
df["Description"] = df["Description"].astype("string")
df["Country"] = df["Country"].astype("string")
df.info()
```

```python
# Convertir fechas
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df.head(10)
```

```python
print(f"Tengo {df.duplicated().sum()} filas duplicadas")
```

```python
columnas_duplicadas=df.columns[df.columns.duplicated(keep=False)]
if any(columnas_duplicadas):
    print(f"Las columnas duplicadas son {list(columnas_duplicadas)}")
else:
    print("No hay columnas duplicadas en el DF")
```

```python
# Identificación de filas duplicadas. Creación de un df con los duplicados
df[df.duplicated()].head(10)
```

Podemos ver que los duplicados no son duplicados reales, ya pueden repetirse varios campos, sin embargo campos importantes como description cambian y esto ya me dice que cada registro es diferente uno del otro

```python
df = df.dropna(subset=["CustomerID"])
df = df.dropna(subset=["Description"])
df.info()
```

```python
# Filtrar devoluciones
df = df[~df["InvoiceNo"].str.startswith("C", na=False)]
df.info()
```

```python
# Ver valores unicos de la columna InvoiceNo
print(f"Tengo {df['InvoiceNo'].nunique()} valores únicos en la columna InvoiceNo")
```

```python
# crear métricas útiles
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]
df["year"] = df["InvoiceDate"].dt.year
df["month"] = df["InvoiceDate"].dt.month
df["day"] = df["InvoiceDate"].dt.day
df["hour"] = df["InvoiceDate"].dt.hour
df.info()
```

```python
df.head(10)
```

Metricas útiles:
Revenue per Invoice: df.groupby("InvoiceNo")["TotalPrice"].sum()
df["items_per_invoice"] = df.groupby("InvoiceNo")["Quantity"].transform("sum")
Unique products per Invoice: df["unique_products"] = df.groupby("InvoiceNo")["StockCode"].transform("nunique")
Metrica por cliente (util para segmentación): df["customer_total_spent"] = df.groupby("CustomerID")["TotalPrice"].transform("sum")
Frecuencia de compra: df["customer_orders"] = df.groupby("CustomerID")["InvoiceNo"].transform("nunique")
